```python
import random
import time

# Define the maze grid
maze = [
    ["#", "#", "#", "#", "#", "#", "#", "#"],
    ["#", "P", ".", ".", "#", ".", "E", "#"],
    ["#", "#", "#", ".", "#", ".", "#", "#"],
    ["#", ".", "G", ".", ".", ".", ".", "#"],
    ["#", "#", "#", "#", "#", "#", "#", "#"]
]

# Function to display the maze
def print_maze():
    for row in maze:
        print(" ".join(row))
    print("\n")

# Function to find the position of an element (P, G, E)
def find_position(symbol):
    for i in range(len(maze)):
        for j in range(len(maze[i])):
            if maze[i][j] == symbol:
                return (i, j)
    return None

# Function to move the player
def move_player(direction):
    global maze
    position = find_position("P")
    if position is None:
        return
    px, py = position
    new_x, new_y = px, py
    
    if direction == "W" and px > 0 and maze[px - 1][py] != "#": 
        new_x -= 1  # Up
    elif direction == "S" and px < len(maze) - 1 and maze[px + 1][py] != "#": 
        new_x += 1  # Down
    elif direction == "A" and py > 0 and maze[px][py - 1] != "#": 
        new_y -= 1  # Left
    elif direction == "D" and py < len(maze[0]) - 1 and maze[px][py + 1] != "#":
        new_y += 1  # Right
    else:
        print("Unable to move in that direction!")
        return
    
    # Update the player's position
    maze[px][py] = "."
    maze[new_x][new_y] = "P"

# Function to move the ghost towards the player
def move_ghost():
    global maze
    player_position = find_position("P")
    ghost_position = find_position("G")
    
    if player_position is None or ghost_position is None:
        return
    px, py = player_position
    gx, gy = ghost_position
    
    # Determine best movement direction
    move_options = []
    if gx > px and maze[gx - 1][gy] != "#":
        move_options.append((gx - 1, gy))  # Move up
    if gx < px and maze[gx + 1][gy] != "#":
        move_options.append((gx + 1, gy))  # Move down
    if gy > py and maze[gx][gy - 1] != "#":
        move_options.append((gx, gy - 1))  # Move left
    if gy < py and maze[gx][gy + 1] != "#":
        move_options.append((gx, gy + 1))  # Move right
    
    if move_options:
        new_gx, new_gy = random.choice(move_options)
        maze[gx][gy] = "."
        maze[new_gx][new_gy] = "G"

# Game loop
def game_loop():
    print("\U0001F383 Welcome to the Haunted Maze! Escape before the ghost catches you! \U0001F47B")
    print_maze()
    exit_position = find_position("E")
    if exit_position is None:
        print("No exit found! Game cannot proceed.")
        return
    ex, ey = exit_position
    
    while True:
        direction = input("Move (W/A/S/D): ").upper()
        if direction not in ["W", "A", "S", "D"]:
            print("Invalid move! Use W (up), A (left), S (down), D (right)")
            continue
        
        move_player(direction)  # Move player
        move_ghost()  # Move ghost
        print_maze()
        
        # Check win condition
        player_position = find_position("P")
        if player_position is None:
            print("You were caught by the ghost! Game Over!.")
            break
        px, py = player_position
        if px == ex and py == ey:
            print("\U0001F389 You Escaped! \U0001F3C6")
            break
        
        
        
        time.sleep(0.5)  # Small delay for better gameplay

game_loop()
```
